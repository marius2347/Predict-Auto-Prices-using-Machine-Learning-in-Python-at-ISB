```python
# chech if there are missing values (NaN) in 'mileage'
auto[auto['mileage'].isna()].sample(5)
```

```python
def checkFunction():
    if len(auto[(auto['mileage'].notna()) & (auto['mileage'].str.contains('km/kg')) & ((auto['fuel'] == "Diesel") | (auto['fuel'] == "Petrol"))]) == 0: # there are none
        return True
    else:
        return False, len(auto[(auto['mileage'].notna()) & (auto['mileage'].str.contains('km/kg')) & ((auto['fuel'] == "Diesel") | (auto['fuel'] == "Diesel"))])
# the dataFrame does not contain the requirements
checkFunction()
```

```python
# firstly make a new column with the values from km/kg of other fuels

# mask to filter rows containing 'km/kg'
mask = auto['mileage'].str.contains('km/kg', na=False) & auto['mileage'].notna() & ((auto['fuel'] != "Petrol") | (auto['fuel'] != "Diesel"))

# filter the dataframe based on mask
filtered_auto = auto[mask]

# display the result
filtered_auto.sample(5)
```

```python
# not na values
filtered_auto.isna().sample(5)
```

```python
# create a new column 'mileage_numeric' from the filtered DataFrame and extract the float values from mileage to put in it
auto['mileage_numeric'] = filtered_auto['mileage'].str.extract('(\d+.\d+)').astype(float)
auto.notna().sample(5)
```

```python
# the new column with km/kg values from other options fuel
auto['mileage_numeric'].sample(5)
```

```python
# now replace the Nan Values with km/kg and do mean of km/kg values from other fuel name
auto['mileage'] = auto['mileage'].fillna('km/kg')
checkFunction() # there are 103 NaN values replaced with km/kg
```

```python
# # show a sample
auto[(auto['mileage'].str.contains('km/kg')) & ((auto['fuel'] == "Petrol") | (auto['fuel'] == "Diesel"))].sample(5)
```

```python
auto.sample(5) # dataFrame modified
```

```python
# make the average of mileage_numeric
mean_mileage = auto['mileage_numeric'].mean()

# replace the values
condition = (auto['fuel'].isin(["Petrol", "Diesel"])) & (auto['mileage'].str.contains('km/kg'))
auto.loc[condition, 'mileage'] = str(mean_mileage.__round__(2)) + " km/kg"

# remove the useless column mileage_numeric
auto = auto.drop('mileage_numeric', axis = 1)
```

```python
#5. Keep only the numeric part in the mileage, engine and max_power columns.

# remove non-numeric values in the "mileage" column
auto['mileage'] = pd.to_numeric(auto['mileage'].replace('[^\d.]', '', regex=True), errors='coerce')

# remove non-numeric values in the "engine" column
auto['engine'] = pd.to_numeric(auto['engine'].replace('[^\d]', '', regex=True), errors='coerce')

# remove non-numeric values in the "max_power" column
auto['max_power'] = pd.to_numeric(auto['max_power'].replace('[^\d.]', '', regex=True), errors='coerce')

# replace NaN values with 0
auto[['mileage', 'engine', 'max_power']] = auto[['mileage', 'engine', 'max_power']].fillna(0)

auto.sample(5)
```

```python
# 6. Remove the column torque

auto = auto.drop('torque', axis = 1)
auto.sample(5)
```
